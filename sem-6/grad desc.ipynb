```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("housing.txt",delimiter=',',names=['size', 'garbage', 'price'])
x = df['size']/max(df['size'])
y = df['price']/max(df['price'])
```

```python
l = len(df)
```

```python
theta0 = 0.0
theta1 = 0.0
alpha = 0.1
for i in range(1000):
    sum0 = sum1 = 0.0
    for j in range(l):
        sum0 += (theta0 + theta1*x[j] - y[j])
        sum1 += (theta0 + theta1*x[j] - y[j])*x[j]
    theta0 -= (alpha*sum0)/l
    theta1 -= (alpha*sum1)/l
```

```python
func = lambda x1: (theta0 + theta1*x1)
yarr = np.array([func(xi) for xi in x])
```

```python
plt.plot(x,y,'^')
plt.plot(x,yarr)
```
